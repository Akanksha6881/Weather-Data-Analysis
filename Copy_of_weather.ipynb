```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Weather Dataset.csv")
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
df.head()
```

```python
print(df.shape)
print(df.info())
print(df.isnull().sum())
print(df.describe())
print(df['Weather'].unique())
```

```python
temp = df['Temp_C'].values
print(np.mean(temp))
print(np.max(temp))
print(np.min(temp))
```

```python
fog_data = df[df['Weather'] == 'Fog']
weather_count = df['Weather'].value_counts()
monthly_avg_temp = df.resample('M', on='Date/Time')['Temp_C'].mean()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Date/Time'], df['Temp_C'], label="Temperature (°C)")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.title("Temperature Trend Over Time")
plt.legend()
plt.show()
```

```python
plt.hist(df['Wind Speed_km/h'], bins=30, color='skyblue', edgecolor='black')
plt.title("Wind Speed Distribution")
plt.xlabel("Wind Speed (km/h)")
plt.ylabel("Frequency")
plt.show()
```

```python
monthly_avg_temp.plot(kind='line', figsize=(10,5), title="Monthly Avg Temperature")
plt.ylabel("Temperature (°C)")
plt.show()
```
